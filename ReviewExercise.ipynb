# 2) Write a Python Program and perform the following operations ?
### I.	Create a List {1225, 4986, 6789, 7890, 2345, 6783, 0987, 1234, 8765, 3456}

### VII.	Map function
### VIII.	Using zip

```python
import numpy as np
import cProfile
import tracemalloc

def Mylist():
    lst=[1225, 4986, 6789, 7890, 2345, 6783, 987, 1234, 8765, 3456]
    return lst

def Looplist(lst):
    looplst=[]
    for l in lst:
        looplst.append(l)
    return looplst

def Looprange(lst):
    rnglst=[]
    for l in range(len(lst)):
        rnglst.append(lst[l])
    return rnglst

def comp(lst):
  cmplst = [l for l in lst]
  return cmplst

def enu(lst):
  obj = enumerate(lst)
  enlst = list(obj)
  return enlst

def itrnxt(lst):
  iterlst = iter(lst)
  ls=[]
  while True:
    try:
        ls.append(next(iterlst))
    except StopIteration:
        break
    return ls;

def mapfun(lst):
  maplst = list(map(float, lst))
  return maplst

def zipfun(lst):
  lst2 = [44,45,73,25,8,3,5,77,58,64]
  ziplst = zip(lst, lst2)
  finalst = list(ziplst)
  return finalst

def npmodfun(lst):
    nparray = np.array(lst)
    arr2 = nparray * 2
    return arr2

lst=Mylist()

tracemalloc.start()

cProfile.run('Mylist()')
Mylist()
print("This is the memory usage:", tracemalloc.get_traced_memory())


cProfile.run('Looplist(lst)')
Looplist(lst)
print("This is the memory usage:", tracemalloc.get_traced_memory())

cProfile.run('Looprange(lst)')
Looprange(lst)
print("This is the memory usage:", tracemalloc.get_traced_memory())

cProfile.run('comp(lst)')
comp(lst)
print("This is the memory usage:", tracemalloc.get_traced_memory())

cProfile.run('enu(lst)')
enu(lst)
print("This is the memory usage:", tracemalloc.get_traced_memory())

cProfile.run('itrnxt(lst)')
itrnxt(lst)
print("This is the memory usage:", tracemalloc.get_traced_memory())

cProfile.run('mapfun(lst)')
mapfun(lst)
print("This is the memory usage:", tracemalloc.get_traced_memory())

cProfile.run('zipfun(lst)')
zipfun(lst)
print("This is the memory usage:", tracemalloc.get_traced_memory())

cProfile.run('npmodfun(lst)')
npmodfun(lst)
print("This is the memory usage:", tracemalloc.get_traced_memory())


tracemalloc.stop()
```

## 3)	For a List A, B, C, D, E write a python program to compute all the combinations and permutations 

```python
import cProfile

def permutations(lst):
    # Base case: single element list
    if len(lst) == 1:
        return [lst]

    # Recursive case: remove each element, find permutations of the rest, and prepend the removed element
    perms = []
    for i in range(len(lst)):
        rest = lst[:i] + lst[i+1:]
        for p in permutations(rest):
            perms.append([lst[i]] + p)
    return perms

# Define the list
my_list = ['A', 'B', 'C', 'D', 'E']

# Compute all permutations
print("Permutations:")
all_permutations = permutations(my_list)
for perm in all_permutations:
    print(perm)

cProfile.run('permutations(my_list)')
```

```python
import cProfile

def combinations(lst, r):
    # Base case: no more elements to combine
    if r == 0:
        return [[]]

    combs = []
    for i in range(len(lst)):
        rest = lst[i+1:]
        for c in combinations(rest, r-1):
            combs.append([lst[i]] + c)
    return combs

# Define the list
my_list = ['A', 'B', 'C', 'D', 'E']

# Compute all combinations of all lengths
print("\nCombinations of all lengths:")
for r in range(1, len(my_list) + 1):
    all_combinations = combinations(my_list, r)
    for comb in all_combinations:
        print(comb)
cProfile.run('combinations(my_list,r)')
```
